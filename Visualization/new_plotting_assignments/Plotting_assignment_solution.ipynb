```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set()
```

# Wine reviews dataset

```python
wine_data = pd.read_csv('winemag-data_first150k.csv.zip', index_col=0)
wine_data.head()
```

```python
wine_data.dtypes
```

```python
wine_data.isna().sum()
```

```python
# bar chart of 10 most occurring provinces

na_index = wine_data.loc[wine_data['province'].isna()].index
wine_data.drop(index=na_index ,inplace=True)  # drop only the rows where province is NaN
# print(wine_data.isna().sum())  # check if drop is successful

plt.figure(figsize=(13,5))
plt.bar(wine_data['province'].value_counts()[:10].index, wine_data['province'].value_counts()[:10])
plt.xticks(rotation=40)
plt.show()
```

```python
# upgrade previous plot to show percentages
plt.figure(figsize=(13,5))
plt.bar(wine_data['province'].value_counts()[:10].index, wine_data['province'].value_counts()[:10] / len(wine_data) * 100)
plt.xticks(rotation=90)
plt.show()
```

```python
# bar chart of points awarded, x-axis in ascending order

plt.figure(figsize=(13,5))
plt.bar(wine_data['points'].unique(), wine_data['points'].value_counts())
plt.xticks(np.arange(80,101))
plt.show()
```

```python
# line plot of points awarded

plt.figure(figsize=(13,5))
plt.plot(wine_data['points'].value_counts().sort_index().index, wine_data['points'].value_counts().sort_index())
plt.xticks(np.arange(80,101))
plt.show()
```

```python
# area plot of points awared
plt.figure(figsize=(13,5))
plt.stackplot(wine_data['points'].value_counts().sort_index().index, wine_data['points'].value_counts().sort_index())
plt.xticks(np.arange(80,101))
plt.show()

# note: this type of plot becomes interesting when plotting multiple categories see: 
# https://python-graph-gallery.com/stacked-area-plot/
```

```python
# histogram of wine prices below $200
plt.figure(figsize=(13,5))
plt.hist(wine_data.loc[wine_data['price'] < 200, 'price'])
plt.show()
```

```python
# histogram of wine prices without filter
plt.figure(figsize=(13,5))
plt.hist(wine_data['price'].dropna())  # gives a warning without dropna, as price contains NaN values
plt.show()
```
