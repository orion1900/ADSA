## Adding Noise
An Example of the Gaussian Noise

```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
import pandas as pd
%matplotlib inline 

from scipy import signal

window = signal.gaussian(51, std=7)
plt.plot(window)
plt.title(r"Gaussian window ($\sigma$=7)")
plt.ylabel("Amplitude")
plt.xlabel("Sample")
```

Now, we create a signal with noise.

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from scipy import signal
%matplotlib inline

t = np.linspace(0,10,200) # create a time signal
x1 = np.sin(t) # create a simple sine wave
x2 = x1 + np.random.rand(200) # add noise to the signal
y1 = sp.signal.medfilt(x2,21) # Filter the signal (median filter)

# plot the results
plt.subplot(2,1,1)
plt.plot(t,x2,'yo-')
plt.title('input wave')
plt.xlabel('time')
plt.subplot(2,1,2)
plt.plot(y1,'yo-')
plt.title('filtered wave')
plt.xlabel('time')

plt.show()
```

Let's add some noice to the power of the windmill.....

```python
wind_noise = wind + np.random.rand(len(wind))
```

```python
plt.figure()
plt.plot(wind[1:1000]) 
plt.plot(wind_noise[1:1000])   #
```

There is a lot of noise! Even bigger than the signal itself!

```python
wind_noise[1:10]
```

Let's diminish the error margin a bit, cz the random numbers are between 0 and 1, and the wind value is low....
Also, lets have the random numbers negative as wel


```python
wind_noise = wind +(np.random.rand(len(wind)) *2 -1)/100
# A more extensive one, using normally distributed
# mu, sigma = -1, 10
# np.random.normal(mu, sigma, len(wind))
```

```python
plt.figure()
plt.plot(wind[1:1000]) 
plt.plot(wind_noise[1:1000])   #
```

Let's look at the power, and do some denoising / smoothing

```python
power_noise = 0.15*wind_noise**2
power_noise_med = sp.signal.medfilt(power_noise,21) # Filter the signal (median filter)
plt.plot(power_noise[1:500])
plt.plot(power_noise_med[1:500])
plt.plot(power[1:500])
```

```python
x=5 # Size of the moving window. This is the number of observations used for calculating the statistic.
pd_power_noise = pd.DataFrame(power_noise)
power_noise_rolling  = pd_power_noise[0].rolling(x).mean()
plt.plot(power_noise[0:500])
plt.plot(power[0:500])
plt.plot(power_noise_rolling[0:500])

# You can play with the rolling window!
```

An other way of smoothing...

```python
import scipy.signal as signal

# First, design the Buterworth filter
N  = 3    # Filter order
Wn = 0.1 # Cutoff frequency
B, A = signal.butter(N, Wn, output='ba')
smooth_data = signal.filtfilt(B,A, power_noise[0:500])
plt.plot(power_noise[0:500],'r-')
plt.plot(smooth_data[0:500],'b-')
plt.show()
```
